```python
for i in df.columns:
    df.rename(columns={i:i.strip()},inplace=True)

df.columns
```

```python
df = df[~(df['FWI']=='fire   ')]
```

```python
df['FWI'].unique()
```

```python
df['FWI'] = df['FWI'].astype('float')
```

```python
df.info()
```

```python
df
```

```python
df['Classes'] = df['Classes'].str.strip()
```

```python
dummies = pd.get_dummies(df['Classes'])
dummies
```

```python
df.drop(columns=['Classes'],inplace=True)
```

```python
df = pd.concat([df,dummies],axis=1)
```

```python
df
```

```python
df.corr()
```

```python
col_list = list(df.columns)
col_list
```

```python
col_list.remove('Temperature')
```

```python
x = df[col_list]
y = df['Temperature']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
lr = LinearRegression()
r = Ridge()
l = Lasso()
en = ElasticNet()
```

```python
# model fit
lr.fit(x_train,y_train)
r.fit(x_train,y_train)
l.fit(x_train,y_train)
en.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
r_pred = r.predict(x_test)
l_pred = l.predict(x_test)
en_pred = en.predict(x_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# residuals

lr_residual = y_test - lr_pred
r_residual = y_test - r_pred
l_residual = y_test - l_pred
en_residual = y_test - en_pred
```
